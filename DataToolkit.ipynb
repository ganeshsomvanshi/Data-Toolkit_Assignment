```python
#Q3 Explain the following termsg
#1. The difference in np.array, np.asarray and np.asanyarray 
#2. The difference between Deep copy and shallow copy.
'''
Differences between np.array, np.asarray, and np.asanyarray
np.array():
Purpose: Creates a new array from any object exposing the array interface (like lists or tuples).
Behavior: Always creates a new array, regardless of the input type. If the input is already an array, it makes a copy of it by default (unless specified otherwise).
Usage: Typically used when you want to ensure a new array is created, regardless of the input.
'''
import numpy as np

a = np.array([1, 2, 3])
b = np.array(a)  # b is a new copy of a

'''
np.asarray():
Purpose: Converts the input to an array but does not create a new array if the input is already an array of the same dtype.
Behavior: If the input is already a NumPy array, it returns that array without copying. If the input is a nested sequence, it will create a new array.
Usage: Useful when you want to ensure the input is an array without unnecessary copying.
'''
a = np.array([1, 2, 3])
b = np.asarray(a)   # b is a reference to a, no new copy is made

'''
np.asanyarray():
Purpose: Similar to np.asarray(), but it allows for subclasses of ndarray (like masked arrays or matrix).
Behavior: It will return a subclass if the input is an array subclass. If it’s not an array-like, it converts it into a regular ndarray.
Usage: Useful when you want to retain subclass information if the input is already a subclass of an array.
'''
class MyArray(np.ndarray):
    pass

a = MyArray([1, 2, 3])
b = np.asanyarray(a)

'''
Differences between Deep Copy and Shallow Copy
Shallow Copy:

Definition: A shallow copy creates a new object but inserts references into it to the objects found in the original.
Behavior: Changes made to mutable objects in the shallow copy will reflect in the original object since they share the same references.
Usage: Use shallow copies when you want a new container object, but you don’t mind sharing the elements within it.
'''
import copy

original = [[1, 2, 3], [4, 5, 6]]
shallow_copied = copy.copy(original)  # Creates a shallow copy

shallow_copied[0][0] = 99
print(original)

'''
Deep Copy:

Definition: A deep copy creates a new object and recursively adds copies of nested objects found in the original.
Behavior: Changes made to the deep copy do not affect the original object, as they do not share references.
Usage: Use deep copies when you want to completely clone an object and ensure that changes in the copy do not affect the original.
'''

import copy

original = [[1, 2, 3], [4, 5, 6]]
deep_copied = copy.deepcopy(original)  # Creates a deep copy

deep_copied[0][0] = 99
print(original) 
```

```python
#Q5 Create a NumPy array with random integers between 1 and 10 of shape (5,6 ). After creating the array 
'''
perform the following operations:

 a)Extract all even integers from array.

 b)Extract all odd integers from array.

'''

import numpy as np

# Create a 5x6 array with random integers between 1 and 10
random_array = np.random.randint(1, 11, size=(5, 6))

print("Original Array:\n", random_array)

# a) Extract all even integers from the array
even_integers = random_array[random_array % 2 == 0]
print("\nEven Integers:\n", even_integers)

# b) Extract all odd integers from the array
odd_integers = random_array[random_array % 2 != 0]
print("\nOdd Integers:\n", odd_integers)
```

```python
#Q14 
'''
Create a time-series dataset in a Pandas DataFrame with columns: 'Date', 'Temperature', 'Humidity' and
Perform the following tasks using Matplotlib: a) Plot the 'Temperature' and 'Humidity' on the same plot with different y-axes (left y-axis for 'Temperature' and 
right y-axis for 'Humidity').

b) Label the x-axis as 'Date'.

c) Set the title of the plot as 'Temperature and Humidity Over Time'.
'''
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Create a date range
dates = pd.date_range(start='2023-01-01', periods=100, freq='D')

# Generate random temperature values between 20 and 35
temperature = np.random.uniform(20, 35, size=len(dates))

# Generate random humidity values between 30 and 70
humidity = np.random.uniform(30, 70, size=len(dates))

# Create a DataFrame
df = pd.DataFrame({'Date': dates, 'Temperature': temperature, 'Humidity': humidity})

# Set the Date as the index
df.set_index('Date', inplace=True)

# a) Plot Temperature and Humidity with different y-axes
fig, ax1 = plt.subplots(figsize=(10, 6))

# Plot Temperature
ax1.set_xlabel('Date')
ax1.set_ylabel('Temperature (°C)', color='tab:red')
ax1.plot(df.index, df['Temperature'], color='tab:red', label='Temperature', linewidth=2)
ax1.tick_params(axis='y', labelcolor='tab:red')

# Create a second y-axis for Humidity
ax2 = ax1.twinx()
ax2.set_ylabel('Humidity (%)', color='tab:blue')
ax2.plot(df.index, df['Humidity'], color='tab:blue', label='Humidity', linewidth=2)
ax2.tick_params(axis='y', labelcolor='tab:blue')

# b) Label the x-axis
ax1.set_xlabel('Date')

# c) Set the title of the plot
plt.title('Temperature and Humidity Over Time')

# Show the plot
plt.tight_layout()  # Adjust layout to prevent overlap
plt.show()
```

```python
#Q16 
'''
Create a NumPy array data containing 1000 samples from a normal distribution. Perform the following
tasks using Matplotlib: a) Plot a histogram of the data with 30 bins.

b) Overlay a line plot representing the normal distribution's probability density function (PDF). 

c) Label the x-axis as 'Value' and the y-axis as 'Frequency/Probability'.

d) Set the title of the plot as 'Histogram with PDF Overlay'.
'''

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Create a NumPy array with 1000 samples from a normal distribution
data = np.random.normal(loc=0, scale=1, size=1000)  # mean=0, std=1

# a) Plot a histogram of the data with 30 bins
plt.hist(data, bins=30, density=True, alpha=0.6, color='g', label='Histogram')

# b) Overlay a line plot representing the normal distribution's PDF
xmin, xmax = plt.xlim()  # Get the limits of x-axis
x = np.linspace(xmin, xmax, 100)  # Create an array of 100 points from xmin to xmax
p = norm.pdf(x, loc=0, scale=1)  # Calculate the PDF

plt.plot(x, p, 'k', linewidth=2, label='Normal Distribution PDF')

# c) Label the x-axis and the y-axis
plt.xlabel('Value')
plt.ylabel('Frequency/Probability')

# d) Set the title of the plot
plt.title('Histogram with PDF Overlay')

# Show legend
plt.legend()

# Show the plot
plt.show()
```

```python
#Q17 Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the origin (quadrants), add a legend, label the axes, and set the title as 'Quadrant-wise Scatter Plot'.

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Generate random data
np.random.seed(0)  # For reproducibility
x = np.random.randn(100)
y = np.random.randn(100)

# Create a DataFrame
data = pd.DataFrame({'x': x, 'y': y})

# Define a function to assign quadrants
def get_quadrant(row):
    if row['x'] > 0 and row['y'] > 0:
        return 'Quadrant 1'
    elif row['x'] < 0 and row['y'] > 0:
        return 'Quadrant 2'
    elif row['x'] < 0 and row['y'] < 0:
        return 'Quadrant 3'
    else:
        return 'Quadrant 4'

# Apply the function to get quadrants
data['Quadrant'] = data.apply(get_quadrant, axis=1)

# Create the scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='x', y='y', hue='Quadrant', palette='viridis', style='Quadrant', s=100)

# Add legend, labels, and title
plt.legend(title='Quadrants')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Quadrant-wise Scatter Plot')
plt.axhline(0, color='grey', lw=0.5, ls='--')
plt.axvline(0, color='grey', lw=0.5, ls='--')
plt.grid(True)
plt.show()
```

```python
#Q18 With Bokeh, plot a line chart of a sine wave function, add grid lines, label the axes, and set the title as 'Sine Wave Function'.
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np

# Prepare the data
x = np.linspace(0, 2 * np.pi, 100)  # 100 points from 0 to 2π
y = np.sin(x)

# Create a new plot with a title and axis labels
p = figure(title="Sine Wave Function", x_axis_label='x', y_axis_label='sin(x)', 
           width=800, height=400)

# Add grid lines
p.grid.grid_line_color = 'gray'
p.grid.grid_line_alpha = 0.5

# Add the sine wave line
p.line(x, y, line_width=2, color='blue', legend_label='sin(x)')

# Show the legend
p.legend.location = "top_left"

# Output the plot
output_notebook()  # Use this if running in a Jupyter notebook
show(p)
```

```python
#Q19 Using Bokeh, generate a bar chart of randomly generated categorical data, color bars based on their values, add hover tooltips to display exact values, label the axes, and set the title as 'Random Categorical Bar Chart'.
import random
from bokeh.io import output_notebook, show
from bokeh.plotting import figure
from bokeh.models import ColumnDataSource, HoverTool
from bokeh.transform import factor_cmap
from bokeh.palettes import Viridis256

# Display plots in the notebook
output_notebook()

# Generate random categorical data
categories = ['Category A', 'Category B', 'Category C', 'Category D', 'Category E']
values = [random.randint(1, 100) for _ in categories]

# Create a ColumnDataSource
source = ColumnDataSource(data=dict(categories=categories, values=values))

# Create a figure
p = figure(x_range=categories, title="Random Categorical Bar Chart",
           toolbar_location=None, tools="")

# Add bars with color mapping based on values
p.vbar(x='categories', top='values', width=0.9, source=source,
       line_color='white', fill_color=factor_cmap('categories', palette=Viridis256, factors=categories))

# Add hover tooltips
hover = HoverTool()
hover.tooltips = [("Category", "@categories"), ("Value", "@values")]
p.add_tools(hover)

# Label the axes
p.xaxis.axis_label = "Categories"
p.yaxis.axis_label = "Values"

# Remove grid lines
p.xgrid.grid_line_color = None
p.y_range.start = 0

# Show the plot
show(p)
```

```python
#Q20 Using Plotly, create a basic line plot of a randomly generated dataset, label the axes, and set the title as'Simple Line Plot'

import plotly.express as px
import pandas as pd
import numpy as np

# Generate a random dataset
np.random.seed(42)  # For reproducibility
x = np.arange(1, 101)
y = np.random.randn(100).cumsum()

# Create a DataFrame
df = pd.DataFrame({'x': x, 'y': y})

# Create a line plot
fig = px.line(df, x='x', y='y', title='Simple Line Plot')

# Label the axes
fig.update_layout(xaxis_title='X Axis', yaxis_title='Y Axis')

# Show the plot
fig.show()
```

```python
#Q21  Using Plotly, create an interactive pie chart of randomly generated data, add labels and percentages, setthe title as 'Interactive Pie Chart'.

import plotly.express as px
import pandas as pd
import numpy as np

# Generate random data
np.random.seed(42)  # For reproducibility
categories = ['Category A', 'Category B', 'Category C', 'Category D', 'Category E']
values = np.random.randint(1, 100, size=len(categories))

# Create a DataFrame
df = pd.DataFrame({'categories': categories, 'values': values})

# Create a pie chart
fig = px.pie(df, values='values', names='categories', title='Interactive Pie Chart',
             labels={'categories': 'Category', 'values': 'Value'},
             hover_data={'values': ':.2f'})

# Add percentages to the labels
fig.update_traces(textinfo='label+percent', insidetextorientation='radial')

# Show the plot
fig.show()
```
